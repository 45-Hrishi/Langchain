##### **2. Why we need Vector DB ?**

Vector databases are specifically designed to store and manage high-dimensional vector data, making them ideal for use cases like word embeddings, image embeddings, and other machine learning-based representations. While SQL and NoSQL databases are efficient for traditional structured or semi-structured data (e.g., text, numbers, JSON documents), they are not optimized for handling vectors in the way vector databases are.

###### **2.1 Why Use Vector Databases Over SQL/NoSQL Databases for Storing Vectors?**

| **Feature**                     | **SQL/NoSQL Databases** | **Vector Databases**                              |
|----------------------------------|-------------------------|---------------------------------------------------|

###### **2.2 Vector Indexes and its Importance**

A **vector index** is a data structure or algorithm used to efficiently search and retrieve high-dimensional vector data based on similarity **(such as cosine similarity or Euclidean distance)**. It enables fast approximate nearest neighbor (ANN) searches by indexing the vectors, allowing you to find vectors that are closest to a given query vector without scanning the entire dataset.

This is particularly useful in machine learning applications where you work with embeddings like **word embeddings, image embeddings, or user behavior embeddings.** These embeddings are high-dimensional vectors (e.g., 100 to 1000 dimensions) that represent complex features, and finding the closest vectors (most similar) is a common operation in tasks such as recommendation systems, image retrieval, and semantic search

Vector indexes are important because searching through millions or billions of high-dimensional vectors for similar items is computationally expensive. Performing a brute-force search, which involves comparing every vector with the query vector, is inefficient and slow. Vector indexes optimize this process by organizing the data in a way that allows for fast lookups and similarity comparisons.

**Common vector index algorithms :**
- **HNSW** (Hierarchical Navigable Small World)
- **FAISS** (Facebook AI Similarity Search)
- **Annoy** (Approximate Nearest Neighbors Oh Yeah)
- **LSH** (Locality Sensitive Hashing)
