##### **2. PDF Handling**

```python
loader = PyPDFLoader("/content/pdfs/attention is all you need.pdf")

extracted_text = []
pages = len(loader.load())
for page in range(pages):
  extracted_text.append(loader.load()[page])

# extracted_text
```

##### **3. Tokenization of text**

For different different problem statements, we have to change the splitting of the text based on the problem statement we have.

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,
                                               chunk_overlap=20)
```

```python
text_chunks = text_splitter.split_documents(extracted_text)
```

```python
print(text_chunks[10].page_content)
```

##### **5. Embeddings**

```python
from langchain_community.embeddings import HuggingFaceInstructEmbeddings
import torch
model_name = "hkunlp/instructor-large"
DEVICE = "cuda:0" if torch.cuda.is_available() else "cpu"
embedding = HuggingFaceInstructEmbeddings(model_name=model_name, model_kwargs={"device": DEVICE})
```

The vector size is 1024 means it is collecting 1024 features for a single line and based on content it categorize the features.

##### **6. Pincone Set up**

```python
PINCONE_API_KEY = os.environ.get('PINCONE_API_KEY')
PINCONE_API_ENV = os.environ.get('PINCONE_API_ENV')
```

```python
PINCONE_API_KEY
```

```python
pc = Pinecone(api_key=PINCONE_API_KEY)
```

The index_name should be in lowercase

```python
index_name = "tutorialindex"

pc.create_index(
    name=index_name,
    dimension=768,
    metric="cosine",
    spec=ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    )
)
```

```python
index = pc.Index(index_name)
```

##### **7. Vector Store and Querying**

```python
vector_store = PineconeVectorStore(index=index, embedding=embedding)
```

```python
uuids = [str(uuid4()) for _ in range(len(text_chunks))]
```

```python
len(uuids)
```

Adding into the vectorstores

```python
vector_store.add_documents(documents=text_chunks,ids=uuids)
```

Query vectorstore

```python
query_documents = vector_store.similarity_search(query="What is the BLEU score for transformer architecture?",k=2)
```

```python
text = ""
for q_doc in query_documents:
  text += q_doc.page_content
```

```python
text
```

##### **8. MistralAI LLM and Retriever**

```python
llm = ChatMistralAI(
    model="mistral-large-latest",
    # model = "Mistral-7B-Instruct-v0.3",
    temperature=0.3,
    api_key = os.environ["MISTRAL_API_KEY"]
)
```

```python
llm
```

A retriever is an interface that returns documents given an unstructured query.Retrievers accept a string query as input and return a list of Document's as output.

```python
qa = RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=vector_store.as_retriever())
```

```python
qa
```

```python
query = "What is the BLEU score for transformer architecture?"
print(qa.invoke(query))
```

##### **9. Simple QA code for testing**

```python
import sys
while True:
  user_input = input("Ask the question : ")
  if user_input.strip() == "exit" or user_input.strip() == "quit":
    time.sleep(1)
    print("Exiting")
    print(time.sleep(5))
    sys.exit()
  if user_input == "":
    continue
  answer = qa.invoke({"query":user_input})
  print(answer["result"])
  print(f"{'*'*200}")
```
